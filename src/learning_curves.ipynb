```python
df = pd.read_csv('../callbacks/losses.csv')

n_cols = len(df.columns)
fig, axs = plt.subplots(figsize=(15, 5))
for i, column in enumerate(df.columns):
    if i == 0: continue
    if i == 4:
        ax2 = axs.twinx()
        ax2.plot(df.index, df[column], label=column, color='red')
    plt.plot(df.index, df[column], label=column)
    print(column)
plt.legend(loc='upper right')
plt.tight_layout()
plt.xlim(-1,1000)
axs.xaxis.set_major_locator(MultipleLocator(100))
axs.xaxis.set_minor_locator(MultipleLocator(50))
axs.yaxis.set_major_locator(MultipleLocator(0.2))
axs.yaxis.set_minor_locator(MultipleLocator(0.1))
axs.tick_params(which='major', length=5,width=1)
axs.tick_params(which='minor', length=3,width=1)
plt.show()
```
